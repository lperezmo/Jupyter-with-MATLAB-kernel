### How to use functions inside of jupyter

If you try to type in functions directly the kernel will throw an error. 

Save the functions to a file by adding the follow line at the top:

```
%%file NAME-OF-FUNCTION.m
```

````matlab
%%file logistic_growth.m
function N = logistic_growth(No, r, K, times)
    N = K./(1+(K/No - 1)*exp(-r*times));
end
````

````matlab
% Loading time series data from file
fid  = fopen('data/logistic_model_hypothetical_data.tsv');
data = cell2mat(textscan(fid, '%f\t%f','CommentStyle', '#','CollectOutput', true));
fclose(fid);

% print size of loaded data set
fprintf('The loaded data has %d rows and %d columns\n',size(data,1),size(data,2));
````

````matlab
% plot the data
times = data(:,1);
popsizes = data(:,2);
plot(times, popsizes, 'Color', 'blue', 'LineWidth',2)
xlabel('Time');
ylabel('Pop. size');
````

````matlab
% plot exponential function on top
hold on;
X = linspace(0,20,100);
Y = 10 * exp(0.3 * X);
plot(X, Y, 'Color', 'red');
hold off;
Ylogistic = logistic_growth(10, 0.5, 1000, X);
hold on;
plot(X, Ylogistic, 'Color', 'black', 'LineWidth', 2);
hold off;
````
