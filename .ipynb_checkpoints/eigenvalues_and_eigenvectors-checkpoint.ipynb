# Overview
`Markdown` cells allow you to leave comments throughout the notebook. They support latex notation (i.e. $\omega, \alpha, \beta$) and the whole notebook can be exported as a `pdf` at the end.

```octave
% Parameter values
tauone = 0.5;
tautwo = 20;
tauthree = 5;
r = 50;
deltaone = 2;
deltatwo = 0.5;
deltathree = 1;
deltafour = 0.05;

% Times for simulation of trajectory
max_time = 50;
Ntimes = 500;
times = linspace(0, max_time, Ntimes);

% Initial values for simulation of trajectory
xone0 = 100;
xtwo0 = 0;
xthree0 = 0;
xfour0 = 0;
Xstart = [xone0; xtwo0; xthree0; xfour0];

% Matrix
G = [-deltaone-1./tauone, 0, 0, r; 1./tauone, -deltatwo-1./tautwo, 0, 0;...
    0, 1./tautwo, -deltathree-1./tauthree, 0; 0, 0, 1./tauthree, -deltafour];
Ndim = size(G,1);
```

```octave
% Calculating eigenvalues and eigenvectors
[V,D] = eig(G);
lambdas = diag(D);

% To make sure the third eigenvalue is not the same as the second
V(:,3) = V(:,3)*i;


% Printing out eigenvalues and eigenvectors in LaTeX-ready form
for i = 1:Ndim
    fprintf('\\lambda_%d & = %f%+fi \\\\ \n',i,real(lambdas(i)),imag(lambdas(i)));
end

fprintf('Eigenvectors are:\n');
for i=1:Ndim
    fprintf('Eigenvector %g\n', i)
    for k=1:Ndim
        fprintf('%f%+fi \\\\ \n', real(V(k,i)), imag(V(k,i)))
    end
end
```

```octave
% Calculating trajectories

trajectory = zeros([Ntimes,Ndim]);
for t=1:Ntimes
    trajectory(t,:) = expm(times(t)'*G) * Xstart;
end
```

```octave
% Plot all components of the trajectory

tiledlayout(2,2);
for i=1:Ndim
    nexttile;
    plot(times, trajectory(:,k), 'LineWidth', 2);
    title(sprintf('Simulated trajectory, %d',i));
    xlabel('Time');
    ylabel('Simulated trajectory');
end
```
